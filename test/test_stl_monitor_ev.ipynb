```python
import stlrom as sr
import numpy as np

sr.Signal().set_BigM(5)

d1 = sr.STLDriver()
d1.parse_string(
"""
signal x
mu := x[t]>0
phi_ev0 := ev_[0, 2.]  ( x[t]>0 )
phi_ev1 := ev_[1., 2.] ( x[t]>0 )
phi_alw0 := alw_[0, 2]   ( x[t]>0 )
phi_alw1 := alw_[1, 2]   ( x[t]>0 )

""")
d1.add_sample([0., 0.1])  
d1.add_sample([0.5, 2])
d1.add_sample([1.5, 1])
d1.add_sample([3,2])

mu = d1.get_monitor('mu')
tr_mu = mu.formula   
st_mu = tr_mu.get_formula_string()
mu.set_eval_time(0,5)
mu.eval_rob()
print('z:')
print(tr_mu.z)
print('z_low:')
print(tr_mu.z_low)
print('z_up:')
print(tr_mu.z_up)
mu.plot_signal('x')
ax = tr_mu.z.plot(label='rob_estimate')
tr_mu.z_up.plot(label= 'rob_up',ax=ax)
tr_mu.z_low.plot(label= 'rob_low', ax=ax)
```

# phi_ev0

```python
phi_ev0 = d1.get_monitor('phi_ev0')
tr_phi = phi_ev0.formula   
st_phi= tr_phi.get_formula_string()

phi_ev0.eval_rob(0,5)
print('z:')
print(tr_phi.z)
print('z_low:')
print(tr_phi.z_low)
print('z_up:')
print(tr_phi.z_up)
```

```python
ax0 = phi_ev0.plot_signal('x')
tr_phi.z.plot(label= 'phi_estimate', ax=ax0)

ax = tr_mu.z.plot('z_mu')
tr_phi.z.plot(label= 'estimate', ax=ax)
tr_phi.z_low.plot(label= 'z_low', ax=ax)
tr_phi.z_up.plot(label= 'z_up', ax=ax)
ax.set_title(tr_phi.get_formula_string())
```
