```python
import stlrom as sr
import numpy as np

sr.Signal().set_BigM(5)

d1 = sr.STLDriver()
d1.parse_string(
"""
signal x
mu := x[t]>0
phi_ev0 := ev_[0, 2.]  ( x[t]>0 )
phi_ev1 := ev_[1., 2.] ( x[t]>0 )
phi_alw0 := alw_[0, 2]   ( x[t]>0 )
phi_alw1 := alw_[1, 2]   ( x[t]>0 )

""")
d1.add_sample([0., 0.1])  
d1.add_sample([0.5, 2])
d1.add_sample([1.5, 1])
d1.add_sample([3,2])
```

```python
mu = d1.get_monitor('mu')
st_mu = mu.formula.get_formula_string()
mu.set_eval_time(0,5)
[rob_mu, rob_mu_low, rob_mu_up] = mu.eval_online_rob()
print('rob_mu:\n', rob_mu, '\nrob_low:\n',rob_mu_low,'\nrob_up:\n',rob_mu_up)
```

```python
mu.plot_signal('x')
ax = rob_mu.plot(label='rob_mu_estimate')
rob_mu_up.plot(label= 'rob_mu_up',ax=ax)
rob_mu_low.plot(label= 'rob_mu_low', ax=ax)
```

# phi_ev0

```python
phi_ev0 = d1.get_monitor('phi_ev0')
[rob, rob_low, rob_up] = phi_ev0.eval_online_rob(0,5)
print('rob:\n', rob, '\nrob_low:\n',rob_low,'\nrob_up:\n',rob_up)
```

```python
ax0 = phi_ev0.plot_signal('x')
rob.plot(label= 'phi_estimate', ax=ax0)

ax = rob_mu.plot('rob_mu')
rob.plot(label= 'rob_estimate', ax=ax)
rob_low.plot(label= 'rob_low', ax=ax)
rob_up.plot(label= 'rob_up', ax=ax)
ax.set_title(phi_ev0.formula.get_formula_string())
```
